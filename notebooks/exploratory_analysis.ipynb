```python
# Load libraries
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load endothelial data
adata = sc.read_h5ad("../data/processed/endothelial_cells.h5ad")
print(f"Endothelial cells: {adata.n_obs}")
print(f"Genes: {adata.n_vars}")
```

```python
# Show tissue distribution
print("\nTissue category distribution:")
print(adata.obs['tissue_category'].value_counts())
```

```python
# Basic visualization - PCA
sc.tl.pca(adata)
sc.pl.pca(adata, color='tissue_category', title='Endothelial Cells: Tumor vs Normal')
```

```python
# Cell 5: Top variable genes
sc.pp.highly_variable_genes(adata, n_top_genes=2000)
sc.pl.highly_variable_genes(adata)
```

```python
# Cell 6: UMAP clustering
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['tissue_category', 'Sample_Origin'])
```

```python
# Cell 7: Preliminary gene expression
# known angiogenesis markers ? 
markers = ['VEGFA', 'PECAM1', 'VWF', 'CDH5', 'FLT1']
sc.pl.dotplot(adata, markers, groupby='tissue_category')
```
